Anatomy Of A Plot<br>
The Canvas / Figure is the overall window or page that everything is drawn on. <br>
It's the top-level component of all the ones that you will consider in the <br>
following points. You can create multiple independent Figures. <br>
A Figure can have several other things in it, such as a subtitle, which is <br>
a centered title to the figure. You'll also find that you can add a legend <br>
and color bar to your Figure.<br>
Axes - most plotting ocurs on an Axes. The axes is the area on which the <br>
data is plotted and that can have ticks, labels, etc. associated with it. <br>
This explains why Figures can contain multiple Axes.<br>
Each Plot has an x-axis and a y-axis, which contain ticks, which have major <br>
and minor ticklines and ticklabels. Thereâ€™s also the axis labels, title, and <br>
legend to consider when you want to customize your axes, but also taking into <br>
account the axis scales and gridlines might come in handy.<br>
Spines are lines that connect the axis tick marks and that designate the <br>
boundaries of the data area. In other words, they are the simple black <br>
square that you get to see when you donâ€™t plot any data at all but when you <br>
have initialized the Axes.

import the necessary packages and modules<br>
prepare data

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
print(X)
C = np.cos(X)
S = np.sin(X)
# prepare plot
plt.figure()
plt.plot(X,C)
plt.plot(X,S)
# show plot
plt.show()
```

line chart<br>
import the necessary packages and modules

```python
import matplotlib.style as style
# use style
style.use('ggplot')
# prepare data
x1 = [5,8,10]
y1 = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
# prepare plot
plt.figure()
plt.plot(x1,y1,label='series 1')
plt.plot(x2,y2,label='series 2')
# annotations
plt.title('Graph Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')
# show plot
plt.grid(True,color='k')
# add a legend
plt.legend()
plt.show()
```

```python
import pandas as pd
# prepare data
# create dataframe
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
        'female': [0, 1, 1, 0, 1],
        'age': [42, 52, 36, 24, 73],
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])
print(df)
```

scatter plot using dataframe<br>

```python
plt.figure()
plt.scatter(df.age, df.preTestScore,label='PreTest')
plt.scatter(df.age, df.postTestScore,label='PostTest')
# annotation
plt.title('Graph Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
# show plot
plt.show()
```
